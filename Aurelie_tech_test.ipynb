```python
test_list = ["aBed", "hello", "w ld"] 
```

```python
#!/usr/bin/env python3
import string #https://www.kite.com/python/docs/string.ascii_lowercase

#Exemple : ["aBed", "hello", "w ld"] => [3, 0, 1]

def get_alphabet():
    """Returns list of alphabet
    
    Params:
        None 
    Return:
        alphabet (list) : List of alphabet letters
    """

    alphabet = list(string.ascii_lowercase)
    return alphabet

def lower_all_string(str):
    """Returns lowered string

    Params:
        str (list) : List of strings 
    Return:
        copy_if_string_list (list) : List of strings lowered
    """

    copy_if_string_list = [ word.lower() for word in str ]
    return copy_if_string_list

def get_count_of_index_matches(string_list):
    """Find count of letters having their index matching 
    within a word and the alphabet
    
    Params:
        string_list (list) : List of strings 
    Return:
        count_list (list) : List of matches' count for each string
    """

    assert len(string_list) > 0, 'Votre liste ne comporte pas d\'éléments'

    copy_if_string_list = lower_all_string(string_list)
    
    alphabet = get_alphabet()
    count_list = []

    for string_ in copy_if_string_list:
        count_list.append(0)

        for character in string_:
            if character in alphabet:
                if string_.index(character) == alphabet.index(character):
                    count_list[copy_if_string_list.index(string_)] += 1

    del copy_if_string_list
    return count_list

if __name__ == '__main__':
    counts = get_count_of_index_matches(test_list)
    print(f'Result is {counts}')

```
